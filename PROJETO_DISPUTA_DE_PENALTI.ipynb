```python
import random

times = [
    "São Paulo", "Corinthians", "Palmeiras", "Santos", "River Plate", "Boca Juniors",
    "Flamengo", "Botafogo", "Peñarol", "Racing", "Grêmio", "Internacional", "LDU",
    "Cruzeiro", "Independiente", "Nacional"
]

print("🔥 Bem-vindo à Copa Libertadores! 🔥\nEscolha seu time para disputar a glória:\n")


for i, time in enumerate(times, 1):
    print(f"{i}. {time}")


indice = int(input("\nDigite o número do seu time: ")) - 1


if indice < 0 or indice >= len(times):
    print("Escolha inválida! Reinicie o jogo.")
    exit()

time_usuario = times[indice]


times_disponiveis = times[:indice] + times[indice + 1:]
time_adversario = random.choice(times_disponiveis)

print(f"\n⚽ Você escolheu: {time_usuario}")
print(f"\n🔥 Seu adversário será: {time_adversario}")

placar_usuario = 0
placar_pc = 0

print(f"\n🏆 A grande disputa de pênaltis começa entre {time_usuario} e {time_adversario}! 🏆")

for i in range(5):
    print(f"\n⚽ Cobrança {i+1}/5 ⚽")

    escolha_usuario = int(input("\nEscolha um número de 1 a 5 para chutar: "))
    print("\n🏃🏾 O jogador corre para a bola... torcida apreensiva...")

    escolha_pc = random.randint(1, 5)

    if escolha_usuario == escolha_pc:
        print("🧤 QUE DEFESA! O GOLEIRO SALVA O TIME!")
    else:
        print("🥅 GOOOOOOL! EXPLODE O ESTÁDIO!!")
        placar_usuario += 1

    escolha_pc = random.randint(1, 5)
    escolha_usuario = int(input("\nEscolha um número de 1 a 5 para tentar defender: "))
    print("\n🏃🏾 O adversário se prepara... silêncio total no estádio...")

    if escolha_usuario == escolha_pc:
        print("🧤 DEFESAA ESPETACULAR! O GOLEIRO SALVA O TIME!")
    else:
        print("🥅 GOL DO ADVERSÁRIO! A TORCIDA ADVERSÁRIA VAI À LOUCURA!")
        placar_pc += 1

    print(f"\n📢 Placar parcial: {time_usuario} {placar_usuario} x {placar_pc} {time_adversario}")

print("\nMOMENTO DE TENSÃO... O PLACAR FINAL ESTÁ SENDO REVELADO...")


if placar_usuario > placar_pc:
    print(f"\n🎉 PARABÉNS! {time_usuario} venceu a disputa por {placar_usuario} x {placar_pc}!")
elif placar_usuario < placar_pc:
    print(f"\n😢 INFELIZMENTE, {time_adversario} venceu por {placar_pc} x {placar_usuario}.")
else:
    print("\n🤝 EMPATE! A DISPUTA FOI ACIRRADA!")

    print("\nObrigado por jogar! ⚽🏆")
```
